```python
import pandas as pd

data = {
    "product": ["Beer", "Wine", "Whisky"],
    "revenue": [120, 200, 300]
}

df = pd.DataFrame(data)

df
```

```python
data = {
    "product": ["Beer", "Beer", "Wine", "Whisky", "Wine", "Beer", "Cocktail", "Cocktail"],
    "day": ["Mon", "Tue", "Tue", "Wed", "Thu", "Thu", "Fri", "Sat"],
    "revenue": [120, 150, 200, 300, 180, 130, 400, 500]
}

df = pd.DataFrame(data)

df
```

```python
revenue_by_product = df.groupby("product")["revenue"].sum().sort_values(ascending=False)

revenue_by_product
```

```python
import matplotlib.pyplot as plt

revenue_by_product.plot(kind="bar")

plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
best_product = revenue_by_product.idxmax()
best_value = revenue_by_product.max()

print(f"Top performing product: {best_product} (£{best_value})")
```

```python
print("INSIGHT:")
print("Cocktails are the top-performing product, generating the highest revenue.")
print("Recommendation: Increase focus on cocktail promotion, especially on weekends.")
```

```python
revenue_by_day = df.groupby("day")["revenue"].sum().sort_values(ascending=False)

revenue_by_day
```

```python
revenue_by_day.plot(kind="bar")

plt.title("Revenue by Day")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.show()
```

```python
best_day = revenue_by_day.idxmax()
best_day_value = revenue_by_day.max()

print(f"Best day: {best_day} (£{best_day_value})")
```
